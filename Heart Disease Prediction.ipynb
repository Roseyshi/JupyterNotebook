Note that trying to plot using categorical data will result into an error. We therefore use the columns that the describe function portrays.

```python
plt.figure(2)
plt.subplot(121)
sns.histplot(data['Age'],kde=True)
plt.subplot(122)  #1st row, 2columns, 2nd chart
data['Age'].plot.box(figsize=(15,5))
```
