```python
import pandas as pd
import numpy as np
from sklearn import linear_model
import matplotlib.pyplot as plt
```

```python
data_frame=pd.read_csv(r'C:\Users\lenovo\Desktop\slr.csv')
data_frame
df=data_frame.dropna(axis=1,how='all')
df
```

```python
#lets define x and y
X=df.iloc[:, :-1].values
y=df.iloc[:, 1].values
```

```python
#lets see the values
X
```

```python
y
```

```python
#we split the data into training and testing sets in order to allow us to determine if our model is good enough
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test=train_test_split(X, y, test_size=0.3, random_state=0)
```

```python
#lets see the values together
X_train, X_test, y_train, y_test
```

```python
from sklearn.linear_model import LinearRegression
#make a function for the linear regression (reg)
reg=LinearRegression()
#we then fit the model to the dataset, first to the training set then to the testing set
reg.fit(X_train, y_train)
```

```python
#now we predict
y_predict=reg.predict(X_test)
y_predict
```

```python
#now we display this on a graph
plt.scatter(X_test, y_test, color='magenta')
plt.plot(X_test, reg.predict(X_test), color='blue')
plt.title('Linear Regression of Area vs Price')
plt.xlabel("area of land")
plt.ylabel('price of land')
plt.show()
```
